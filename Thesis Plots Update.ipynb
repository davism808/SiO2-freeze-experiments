```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
file_path = r"C:\Users\dbail\Downloads\12 Week Data.txt"  
data = pd.read_csv(file_path, delimiter='\t')
# Extract columns into numpy arrays
Cruise = data['Cruise'].values
Fridge_One_Week = data['A Fridge'].values
Freezer_One_Week = data['A -20'].values
Cryogenic_One_Week = data['A -80'].values
Fridge_Six_Weeks = data['B fridge'].values
Freezer_Six_Weeks = data['B -20'].values
Cryogenic_Six_Weeks = data['B -80'].values
Fridge_Twelve_Weeks = data['C fridge'].values
Freezer_Twelve_Weeks = data['C -20'].values
Cryogenic_Twelve_Weeks = data['C -80'].values


# Titles for each subplot
titles = ['Fridge vs. Freezer (One Week)', 'Fridge vs. Cryogenic (One Week)',
          'Freezer vs. Cryogenic (One Week)', 'Fridge vs. Freezer (Six Weeks)',
          'Fridge vs. Cryogenic (Six Weeks)', 'Freezer vs. Cryogenic (Six Weeks)',
          'Fridge vs. Freezer (Twelve Weeks)', 'Fridge vs. Cryogenic (Twelve Weeks)',
          'Freezer vs. Cryogenic (Twelve Weeks)']

# Data for each subplot
data_sets = [(Fridge_One_Week, Freezer_One_Week), (Fridge_One_Week, Cryogenic_One_Week), (Freezer_One_Week, Cryogenic_One_Week),(Fridge_Six_Weeks, Freezer_Six_Weeks), (Fridge_Six_Weeks, Cryogenic_Six_Weeks), (Freezer_Six_Weeks, Cryogenic_Six_Weeks),(Fridge_Twelve_Weeks, Freezer_Twelve_Weeks), (Fridge_Twelve_Weeks, Cryogenic_Twelve_Weeks), (Freezer_Twelve_Weeks, Cryogenic_Twelve_Weeks)]

# Function to convert data to float arrays and handle errors
def convert_to_float(data):
    try:
        return np.array(data, dtype=float)
    except ValueError as e:
        print(f"Error converting data to float: {e}")
        return np.array([])

# Convert data sets
data_sets = [(convert_to_float(x), convert_to_float(y)) for x, y in data_sets]

# Initialize figure and axes
fig, axs = plt.subplots(3, 3, figsize=(15, 15))  # 3 rows, 3 columns

# Plotting each subplot
for i, (x_data, y_data) in enumerate(data_sets):
    row = i // 3
    col = i % 3

    # Check if data conversion was successful
    if len(x_data) == 0 or len(y_data) == 0:
        print(f"Data conversion failed for plot {i+1}, skipping.")
        continue

    # Scatter plot
    r_squared = round(np.corrcoef(x_data, y_data)[0, 1]**2, 4)
    axs[row, col].scatter(x=x_data, y=y_data, alpha=0.5, label='$R^{2}$ = ' + str(r_squared))

    # Plot a trend line
    m, b = np.polyfit(x_data, y_data, 1)
    axs[row, col].plot(x_data, m * x_data + b, color='gray')

    # Set axes limits
    axs[row, col].set_ylim(0, 170)
    axs[row, col].set_xlim(0, 170)

    # Set title and legend
    axs[row, col].set_title(titles[i])
    axs[row, col].legend()



# Adjust layout and save
plt.tight_layout()
plt.savefig('3x3_grid_plots_with_R2.png', dpi=300, bbox_inches='tight')
plt.show()
```
